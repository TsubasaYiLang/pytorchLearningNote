```python
import torch
```

```python
import numpy as np
xy = np.loadtxt("./data/diabetes.csv.gz", delimiter=',', dtype=np.float32)
x_data = torch.from_numpy(xy[:, :-1])
y_data = torch.from_numpy(xy[:, [-1]]) # 加上方括号才表示矩阵，直接写[:, -1]拿出来的是一维向量
x_data, y_data
```

![模型结构](./data/model_structure.png)

神经网络的本质：  
寻找一种非线性空间变换函数  


step 1: 准备数据集  
step 2：构建网络模型  
step 3：构造损失函数和优化器  
step 4：训练周期  

```python
class Model(torch.nn.Module):
  def __init__(self):
    super(Model, self).__init__()
    self.linear1 = torch.nn.Linear(8, 6)
    self.linear2 = torch.nn.Linear(6, 4)
    self.linear3 = torch.nn.Linear(4, 2)
    self.linear4 = torch.nn.Linear(2, 1)
    self.sigmoid = torch.nn.Sigmoid()
  
  def forward(self, x):
    x = self.sigmoid(self.linear1(x))
    x = self.sigmoid(self.linear2(x))
    x = self.sigmoid(self.linear3(x))
    x = self.sigmoid(self.linear4(x))
    return x

model = Model()
```

```python
criterion = torch.nn.BCELoss(reduction='mean')
optimizer = torch.optim.SGD(model.parameters(), lr = 0.1)
```

```python
for epoch in range(1000000):
  # Forward
  y_pred = model(x_data)
  loss = criterion(y_pred, y_data)
  print(epoch, loss.item())

  # Backward
  optimizer.zero_grad()
  loss.backward()

  # Update
  optimizer.step()


        
```

```python
if epoch % 100000 == 99999:
    y_pred_label = torch.where(y_pred>=0.5, torch.tensor([1.0]), torch.tensor([0.0]))
    #print(y_pred_label, y_data)
    acc = torch.eq(y_pred_label, y_data).sum().item()/y_data.size(0)
    print("loss = ",loss.item(), "acc = ",acc)
    
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_pred_label, y_data),
```

```python
classification_report(y_pred_label, y_data)
```
